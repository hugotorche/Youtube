```python
people = {"first": ["Corey", "Jane", "John"], 
          "last": ["Schafer", "Doe", "Doe"], 
          "email": ["CoreyMSchafer@gmail.com", "JaneDoe@gmail.com", "JohnDoe@gmail.com"]}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
df
```

```python
df['email']
```

```python
df.set_index('email')
```

```python
df #the set_index didn't change the df 
```

```python
df.set_index('email', inplace=True)
```

```python
df #the set_index did change the df (thanks inplace !)
```

```python
df.index
```

```python
df.loc['CoreyMSchafer@gmail.com']
```

```python
df.loc['CoreyMSchafer@gmail.com', 'last']
```

```python
df.iloc[0] #doesn't work with loc (because of index change)
```

```python
df.reset_index(inplace=True) #reset index (don't forget inplace !)
```
