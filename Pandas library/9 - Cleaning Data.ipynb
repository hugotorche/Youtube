```python
import pandas as pd 
import numpy as np #for the NaN values

people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
#missing values in np.nan and None
df.dropna()
```

```python
df.dropna(axis='index', how='any') #default parameters => same result 

#index => drop each INDEX 
#how => that contains ANY value 
```

```python
df.dropna(axis='index', how='all')
```

```python
df.dropna(axis='columns', how='all') #doesn't change anything here
```

```python
df.dropna(axis='columns', how='any') #change every thing here 
```

```python
df.dropna(axis='index', how='any', subset=['email'])
```

```python
#If you absolutely need both last name and email
df.dropna(axis='index', how='all', subset=['last','email'])
```

```python
#missing values in str format

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
df.dropna(axis='index', how='all', subset=['last','email'])
```

```python
df.isna()
```

```python
#Replace NA values => most relevant for numerical Data 
df.fillna('MISSING')
```

```python
df.fillna('0')
```

```python
#New column => age
df.dtypes #not a method => attribute
```

```python
df['age'].mean() #doesn't work because not int or float
```

```python
type(np.nan)
```

```python
#Change the type
df['age'] = df['age'].astype(float)
df.dtypes
```

```python
df['age'].mean()
```
